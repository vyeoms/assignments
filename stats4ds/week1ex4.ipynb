## Part a

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf(y):
    return ( 1/(y*np.sqrt(2*np.pi)) ) * np.exp( -np.log(y)*np.log(y)*0.5 )

x = np.arange(0.1, 10, 0.1)
y = pdf(x)

plt.plot(x, y)
plt.show()
```

```python
sample = np.random.normal(0, 1, 10000)
plt.hist(sample, bins=300)
plt.show()

plt.hist(np.exp(sample), bins=300)
plt.xlim(0, 10)
plt.show()
```

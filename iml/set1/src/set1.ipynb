```python
import os
from urllib.request import urlretrieve
import numpy as np
import pandas as pd
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GroupKFold
from sklearn.metrics import mean_squared_error

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00514/Bias_correction_ucl.csv"
file = "../assets/Bias_correction_ucl.csv"

if not os.path.exists(file):
    urlretrieve(url, file)
# Read data, drop Date and Next_Tmin and rows with NA values already at this point
nwp = pd.read_csv(file).drop(["Date", "Next_Tmin"], axis=1).dropna()

# Downsample to 1000 rows, stratifying by the station so that the data contains
# about an equal number of measurements from each of the stations
nwp, _ = train_test_split(
    nwp, train_size=1000, random_state=42, stratify=nwp["station"]
)
X = nwp.drop(["Next_Tmax", "station"], axis=1)
y = nwp["Next_Tmax"]
# It is important to permute data randomly, which is why we have shuffle=True
## (even though it is default).
# We stratify again with respect to the stations to have an equal proportion of
## stations in training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=500, random_state=42, shuffle=True, stratify=nwp["station"]
)

# All requested models are implemented by sklearn, luckily
models = [DummyRegressor(), LinearRegression(), SVR(), RandomForestRegressor(), GradientBoostingRegressor()]

# It is a good idea to collect the results in a data frame, one row for a regressor.
res = pd.DataFrame(index=["dummy", "OLS", "SVR", "RF", "XGBoost"])

# Auxiliary function to compute the mean squared error.
# We could easily write the funtion ourselves, but we opt to use the sklearn mean_squared_error
def loss(X_tr, y_tr, X_te, y_te, m):
    return mean_squared_error(y_te, m.fit(X_tr, y_tr).predict(X_te), squared=False)

# Loss on training data, for model trained on training data:
res["train"] = [loss(X_train, y_train, X_train, y_train, m) for m in models]
# Loss on test data, for model trained on training data:
res["test"] = [loss(X_train, y_train, X_test, y_test, m) for m in models]
# Cross-validation loss:
res["cv"] = [
    -cross_val_score(
        m, X_train, y_train, cv=10, scoring="neg_root_mean_squared_error"
    ).mean()
    for m in models
]

# This time, we don't drop the station information to be able to group the data according to it
X = nwp.drop(["Next_Tmax"], axis=1)
y = nwp["Next_Tmax"]
# It is important to permute data randomly, which is why we have shuffle=True
## (even though it is default).
# We stratify again with respect to the stations to have an equal proportion of
## stations in training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=500, random_state=42, shuffle=True, stratify=nwp["station"]
)

stations = X_train["station"]

# We drop the station after getting doing the permutation to avoid training around it
X_train = X_train.drop(["station"], axis=1)
res["station"] = [
    -cross_val_score(
        m, X_train, y_train, groups=stations, cv=GroupKFold(n_splits=24), scoring="neg_root_mean_squared_error"
    ).mean()
    for m in models
]

res
```

"Leave-one-station-out cross-validation" (LOSOCV) is a cross-validation variation that the authors used, where they leave out the measurements from one meteorological station out to use as a validation set. They do this because there are confounding factors that can introduce systematic errors in the models. LOSOCV allows the models to account for this kind of error when learning from the data.

```python
import numpy as np
import numpy.random as npr
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

def polynomial(x):
    return 2 - x + x**2

def sampler(n):
    x = npr.uniform(-3, 3, n)
    return x, polynomial(x) + npr.normal(0, 0.4, n)

X_train, y_train = sampler(10)
X_val, y_val = sampler(10)
X_test, y_test = sampler(1000)

p = [1, 2, 3, 4, 8]

def createPolyModel(degree):
    # The intercept is handled by PolynomialFeatures
    return Pipeline([('poly', PolynomialFeatures(degree=degree)), ('linear', LinearRegression(fit_intercept=False))])

models = [createPolyModel(degree) for degree in p]

# Collect results in a dataframe, one row per degree of polynomial fitted
res = pd.DataFrame(index=["p=1", "p=2", "p=3", "p=4", "p=8"])

# Auxiliary function to compute the mean squared error.
# We could easily write the function ourselves, but we opt to use the sklearn mean_squared_error
def loss(X_tr, y_tr, X_te, y_te, m):
    return mean_squared_error(y_te, m.fit(X_tr, y_tr).predict(X_te), squared=True)

def plotFitModels(models, x, y, degrees=p):
    resolution = 300
    ground_x = np.linspace(-3, 3, resolution)
    for i in range(5):
        poly_coefs = models[i].fit(x.reshape(-1, 1), y).named_steps['linear'].coef_
        poly_fit = np.zeros(resolution)
        for j in range(len(poly_coefs)):
            poly_fit += poly_coefs[j] * ground_x**j
        plt.plot(ground_x, poly_fit)
        plt.scatter(x, y)
        plt.title("Fit with p={}".format(degrees[i]))
        plt.show()
    return

plotFitModels(models, X_train, y_train)

# Loss on training data, for model trained on training data:
res["train"] = [loss(X_train.reshape(-1, 1), y_train, X_train.reshape(-1, 1), y_train, m) for m in models]
# Loss on validation data, for model trained on training data:
res["validation"] = [loss(X_train.reshape(-1, 1), y_train, X_val.reshape(-1, 1), y_val, m) for m in models]
# Loss on test data, for model trained on training data:
res["test"] = [loss(X_train.reshape(-1, 1), y_train, X_test.reshape(-1, 1), y_test, m) for m in models]

X_combo = np.concatenate((X_val, X_train))
y_combo = np.concatenate((y_val, y_train))
# Cross-validation loss:
res["cv"] = [
    -cross_val_score(
        m, X_combo.reshape(-1, 1), y_combo, cv=5, scoring="neg_root_mean_squared_error"
    ).mean()
    for m in models
]

# Loss on test data, for model trained on combined data:
res["combo_test"] = [loss(X_combo.reshape(-1, 1), y_combo, X_test.reshape(-1, 1), y_test, m) for m in models]

res
```

The bias measures "how far the average model is from the optimal classifier"; with high model flexibility, we are able to capture more nuances and complexity about the data with the model. This is why it has a downward trend with relation to the model complexity, since we can approach an "optimal classifier" better if we have higher complexity. High bias means that the model is not specialized enough, and it cannot explain the phenomena we want to capture to an acceptable degree.

On the other hand, the variance tells us "how far a model tends to be from the average model". It measures how well a model can generalize its beliefs about the data on new, unseen data. If we have a highly complex model, we have encoded a lot of importance on the features we have trained it on, so it will usually perform bad on unseen data. This is also why the training error has a downward trend wrt the model complexity; with a highly flexible model, we are making sure that it describes the *training data* with high accuracy, whereas if we had a simpler model there are aspects of the data we would not capture (and at a certain point, this is a good thing, because we don't want noisy measurements to affect our model too much).

The irreducible error is constant for any complexity, as it describes error from the measurement mechanic that has been used to gather data. This may be something like a sensor's precision, noise in the environment where we gather data, or many other reasons. This is not something we can deal with during the modeling stage, hence its name.

The test set error is a proxy for the total error the model will have on unseen data. The test set itself is a set of the data that we ignore during the training stage of the model, and we use it as a representative sample of the population of interest. With all that has been said before, its absolute minimum may theoretically be the irreducible error if the model has exactly 0 bias and 0 variance. In practice, this doesn't happen, with it being offset by some combination of bias and variance. We usually wish to get a model that has the lowest test set error, since this means that the aggregate error that comes from bias and variance is minimized.

```python
import numpy as np
import numpy.random as npr
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures


def polynomial(x):
    return 2 - x + x**2

def sampler(n, ex_4 = False):
    x = npr.uniform(-3, 3, n)
    return x, polynomial(x) + npr.normal(0, 0.4, n)

def createPolyModel(degree):
    # The intercept is handled by PolynomialFeatures
    return Pipeline([('poly', PolynomialFeatures(degree=degree)), ('linear', LinearRegression(fit_intercept=False))])

# True value of the polynomial
f_0 = polynomial(0)

# Degrees of the polynomials to estimate
p = [1, 2, 3, 4, 6, 8]
m = len(p)

# Linear regression models on each degree
models = [createPolyModel(degree) for degree in p]

npr.seed(0)
# Generate 1000 training sets of 10 datapoints
X_train, y_train = sampler(10000)
y0_array = npr.normal(0, 0.4, 1000) + f_0
X_train = X_train.reshape(1000, 10)
y_train = y_train.reshape(1000, 10)

f_hat_arrays = np.zeros( (1000, m) )

for i in range(1000):
    # Get a subset of 10 training points
    X_train_subset = X_train[i]
    y_train_subset = y_train[i]
    
    for j in range(m):
        # This is the intercept term for model `j` on test set `i`: at x=0, the estimation is equal to this
        f_hat_arrays[i][j] = models[j].fit(X_train_subset.reshape(-1, 1), y_train_subset).predict( np.array([0]).reshape(-1, 1) )

# Calculate the expected error of each model
# Collect results in a dataframe, one row per degree of polynomial fitted
res = pd.DataFrame(index=["p=" + str(degree) for degree in p])

res["sq_loss"] = np.mean((y0_array.reshape(1000, 1) - f_hat_arrays)**2, axis=0)

# Expected Bayes error
res["bayes_err"] = np.mean((y0_array - f_0)**2)

expected_estimation = np.mean(f_hat_arrays, axis=0)

print(expected_estimation)

res["sq_bias"] = (expected_estimation - f_0)**2

res["var"] = np.mean( (f_hat_arrays - expected_estimation)**2, axis=0)

res
```

```python
import matplotlib.pyplot as plt

plt.plot(p[:-1], res["sq_loss"][:-1], label="square loss")
plt.plot(p[:-1], res["bayes_err"][:-1], label="Bayes error")
plt.plot(p[:-1], res["sq_bias"][:-1], label="square bias")
plt.plot(p[:-1], res["var"][:-1], label="variance")
plt.legend()
plt.xlabel("Degree of polynomial")
plt.title("Square loss, Bayes error, square bias and variance as functions of p")
plt.show()

plt.plot(p[:-1], res["bayes_err"][:-1] + res["sq_bias"][:-1] + res["var"][:-1], label="Sum of Bayes err, bias and var")
plt.plot(p[:-1], res["sq_loss"][:-1], label="Square loss")
plt.legend()
plt.xlabel("Degree of polynomial")
plt.title("Sum of components and square loss")
plt.show()
```

To confirm that the square loss is indeed equal to the sum of the Bayes error, the squared bias and the variance, we can use the second plot as visual aid. I do not compare the sum directly, at least not in code, since this is prone to numerical error.

Visually we can see that the curves of the square loss and the sum of the error terms pretty much match one-to-one, which confirms that formula given by James et. al

The terms on each polynomial model initially seem to behave as expected, but they have some issues that we will address. The model with the lowest variance should be the polynomial of degree 1, but we can see that this is not the case; this could be because since the model is just the intercept term, which corresponds to the mean of the training sets used, it is very prone to errors. Otherwise, we can see that the polynomial of degree 2 has the lowest variance of all the models, after which the variance just increases, which is what we expected.

The calculated square loss is lowest for the degree 2 polynomial model, which is what we expected since the true source of the data is a degree 2 polynomial. On the other hand, we can see that the Bayes error is relatively close to its theoretical value, which should be around $0.4^2=0.16$ (the variance of the noise term on the data).

Finally, there are some issues with numerical stability. We would expect the bias term to go down as the model complexity increases; however, although the second degree model achieves a very low bias, from the degree 3 polynomial onward we see that the bias increases. On top of this, we can see that all the terms for the degree 8 polynomial blow up with exorbitant values. After a discussion with the professor of the course, we have attributed these issues to numerical stability of the sklearn library. The library does *not* use orthogonal polynomials in the calculation of the regression, which makes it numerically unstable.

### Task a

We have:

$$E[L_{test}] = E\left[\frac{1}{m}\sum_{i=1}^m (\bar{y}_i - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_i ) ^2 \right]$$

Due to linearity:

$$E[L_{test}] = \frac{1}{m}\sum_{i=1}^m E\left[(\bar{y}_i - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_i )^2 \right]$$

Note that $\bar{y}_i$ is IID for all $i \in \mathbb{N}$, as is $\bar{\mathbf{x}_i}$, and $\hat{\beta}$ is a constant w.r.t. the test data. Therefore, for our data, we have:

$$E\left[(\bar{y}_1 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_1 )^2 \right] = E\left[(\bar{y}_2 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_2 )^2 \right] = \cdots = E\left[(\bar{y}_i - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_i )^2 \right]$$

So this sum is just equal to $mE\left[(\bar{y}_1 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_1 )^2 \right]$.

$$\therefore E[L_{test}] = \frac{1}{m}\cdot mE\left[(\bar{y}_1 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_1 )^2 \right] = E\left[(\bar{y}_1 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_1 )^2 \right]$$

### Task b

To show that $L_{test}$ is an unbiased estimator of the generalization error is the same as showing that:

$$E[L_{test}] - \text{generalization error} = 0$$

The generalization error of OLS is given by:

$$E\left[(y_k - \hat{\beta}^{\intercal}\mathbf{x}_k )^2 \right]$$

For some unseen observation $(x_k, y_k)$ drawn from the same distribution as our training data. Since all the data is drawn from the same population, we have that:

$$E\left[(y_k - \hat{\beta}^{\intercal}\mathbf{x}_k )^2 \right] = E\left[(y_1 - \hat{\beta}^{\intercal}\mathbf{x}_1 )^2 \right] = E\left[(\bar{y}_1 - \hat{\beta}^{\intercal}\bar{\mathbf{x}}_1 )^2 \right]$$

Therefore, $L_{test}$ is an unbiased estimator of the generalization error.


### Task c

We have:
$$E[L_{train}] = E\left[ \frac{1}{n}\sum_{i=1}^n (y_i - \hat{\beta}^\intercal \mathbf{x}_{i})^2 \right]$$

Due to linearity:

$$E[L_{train}] = \frac{1}{n}\sum_{i=1}^n E\left[ (y_i - \hat{\beta}^\intercal \mathbf{x}_{i})^2 \right]$$

The parameter vector $\hat{\beta}$ is fitted so that the empirical square loss is minimized. This means that the error in a training datapoint is, at most, the generalization error:

$$E\left[ (y_i - \hat{\beta}^\intercal \mathbf{x}_{i})^2 \right] \leq E\left[(y_k - \hat{\beta}^{\intercal}\mathbf{x}_k )^2 \right]$$

$$\Rightarrow \frac{1}{n}\sum_{i=1}^n E\left[ (y_i - \hat{\beta}^\intercal \mathbf{x}_{i})^2 \right] \leq \frac{1}{n}\sum_{i=1}^n E\left[(y_k - \hat{\beta}^{\intercal}\mathbf{x}_k )^2 \right] = \frac{1}{n} \cdot nE\left[(y_k - \hat{\beta}^{\intercal}\mathbf{x}_k )^2 \right] = E\left[(y_1 - \hat{\beta}^{\intercal}\mathbf{x}_1 )^2 \right]$$

$$\therefore E[L_{train}] \leq E[L_{test}]$$

### Task d

In machine learning, we seek to develop models that can *generalize*: we want a model that can perform well on data it hasn't seen before. The *generalization error* is a metric we use to measure such performance, in other words, the generalization error is a measure of how well a model performs on data it hasn't seen.

From this statement, we can see that the test set loss $L_{test}$ is crucial in determining if a model has been trained well. This means that we need a test set that is representative of the population or distribution of interest, but we are limited by our need to train the model. Typically, models with high complexity need copious amounts of training data to avoid overfitting, so if we have a complex problem that needs such a model to represent it properly, we may face issues in distributing our data between the training set and the testing set.

On the flip side, if our training set is big and is representative of the population, the training set loss $L_{train}$ may be used in conjunction with the testing set loss, since it provides a lower bound on the test loss, though keeping in mind that a low training set loss by itself does *not* tell us if the model will perform well on unseen data.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("../assets/co2lite.csv")

df.head()
```

```python
# Mean of FCO2
fco2_mean = df["FCO2"].mean()
print(f"The mean of FCO2 is: {fco2_mean}")

# Standard error for t-statistic and confidence interval calculations
fco2_mean_se = df["FCO2"].sem()

# t-statistic as given by James et. al
t = fco2_mean/fco2_mean_se
print(f"The t-statistic of the mean of FCO2 is: {t}")

# Studentized CI for the mean
fco2_mean_ci = [fco2_mean - 2*fco2_mean_se, fco2_mean + 2*fco2_mean_se]
print(f"Studentized confidence interval for the FCO2 mean: {fco2_mean_ci}")

plt.hist(df["FCO2"], bins=15)
plt.show()
```

Although the histogram shows that many datapoints are close to 0, we can see that the 50 datapoints have a long tail to the left which can skew the mean to a negative value. Added to this, the 95% confidence interval does not contain 0, and the t-statistic indicates that the mean estimation is about 2.3 standard deviations to the left from 0 (as defined by James et. al). Since there are more than 30 data points, the error starts to approximate a normal distribution, which validates the Studentized confidence interval.

All of the previous indicates, with good confidence, that the mean of the dataset is non-zero.

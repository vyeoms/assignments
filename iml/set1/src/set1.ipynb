It seems like the amount of atmospheric CO2 measured at a height of 33.6 m presents some strong inverse correlation with the levels of UV radiation. This makes sense, since the radiation can break down the CO2 molecules into oxygen ions and carbon monoxide, reducing the CO2 measured. On the other hand, the correlation between UV radiation and the levels of nitrous oxide seems to be lower, which could be explained because the bond between a single nitrogen with an oxygen is stronger than the CO2 bonds, thus requiring more energy to break down.

```python
import os
from urllib.request import urlretrieve
import numpy as np
import pandas as pd
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GroupKFold
from sklearn.metrics import mean_squared_error

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00514/Bias_correction_ucl.csv"
file = "../assets/Bias_correction_ucl.csv"

if not os.path.exists(file):
    urlretrieve(url, file)
# Read data, drop Date and Next_Tmin and rows with NA values already at this point
nwp = pd.read_csv(file).drop(["Date", "Next_Tmin"], axis=1).dropna()

# Downsample to 1000 rows, stratifying by the station so that the data contains
# about an equal number of measurements from each of the stations
nwp, _ = train_test_split(
    nwp, train_size=1000, random_state=42, stratify=nwp["station"]
)
X = nwp.drop(["Next_Tmax", "station"], axis=1)
y = nwp["Next_Tmax"]
# It is important to permute data randomly, which is why we have shuffle=True
## (even though it is default).
# We stratify again with respect to the stations to have an equal proportion of
## stations in training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=500, random_state=42, shuffle=True, stratify=nwp["station"]
)

# All requested models are implemented by sklearn, luckily
models = [DummyRegressor(), LinearRegression(), SVR(), RandomForestRegressor(), GradientBoostingRegressor()]

# It is a good idea to collect the results in a data frame, one row for a regressor.
res = pd.DataFrame(index=["dummy", "OLS", "SVR", "RF", "XGBoost"])

# Auxiliary function to compute the mean squared error.
# We could easily write the funtion ourselves, but we opt to use the sklearn mean_squared_error
def loss(X_tr, y_tr, X_te, y_te, m):
    return mean_squared_error(y_te, m.fit(X_tr, y_tr).predict(X_te), squared=False)

# Loss on training data, for model trained on training data:
res["train"] = [loss(X_train, y_train, X_train, y_train, m) for m in models]
# Los on test data, for model trained on training data:
res["test"] = [loss(X_train, y_train, X_test, y_test, m) for m in models]
# Cross-validation loss:
res["cv"] = [
    -cross_val_score(
        m, X_train, y_train, cv=10, scoring="neg_root_mean_squared_error"
    ).mean()
    for m in models
]

# This time, we don't drop the station information to be able to group the data according to it
X = nwp.drop(["Next_Tmax"], axis=1)
y = nwp["Next_Tmax"]
# It is important to permute data randomly, which is why we have shuffle=True
## (even though it is default).
# We stratify again with respect to the stations to have an equal proportion of
## stations in training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=500, random_state=42, shuffle=True, stratify=nwp["station"]
)

stations = X_train["station"]

# We drop the station after getting doing the permutation to avoid training around it
X_train = X_train.drop(["station"], axis=1)
res["station"] = [
    -cross_val_score(
        m, X_train, y_train, groups=stations, cv=GroupKFold(n_splits=24), scoring="neg_root_mean_squared_error"
    ).mean()
    for m in models
]

res
```

### Task b discussion

- The best regressor is between the RandomForest and the selected, additional regressor, XGBoost. Which one is *the best* depends on the error that we wish to select, since RandomForest performs best on the training error as well as the Leave-One-Station-Out Cross-Validation, while XGBoost has lower error in the test set and with 10-fold Cross-Validation. 
- The RMSE on the training data is generally better than on the testing set, with the exceptions being the dummy classifier (which is understandable, since it doesn't learn from the data) and the support vector regressor. The SVR has the worst performance out of all the non-dummy classifiers, and this may be due to the beliefs it encodes about the data; it seems that the regressor isn't able to generate a hyperplane that captures the true structure of the data, at least with the default sklearn parameters, since it has high error for both the training set and the test set. Coupled with the fact that there isn't a significant difference between training and testing errors, the SVR with default parameters is not a good fit for the task at hand, confirmed by the fact that the dummy classifier has similar performance.
- The cross-validation error is lower than in the testing set for the classifiers that aren't the dummy nor the SVR.
- There are some things that could be done with the regressors to help them improve their performance. At this moment, we're just using the default parameters that sklearn brings for each. We can try to fine-tune these parameters to fit the models to the data better, e.g. we could adjust the degree of the polynomial kernel of the SVR which is 3 by default, or tune the number of regressors on each ensemble model.

### Task c discussion

"Leave-one-station-out cross-validation" (LOSOCV) is a cross-validation variation that the authors used, where they leave out the measurements from one meteorological station out to use as a validation set. They do this because there are confounding factors that can introduce systematic errors in the models. For example, they mention that stations 4 and 20 had height differences of 200 m when compared to the others, which induced a bias toward colder temperatures. LOSOCV allows the models to account for this kind of error when learning from the data.
